```python
from control import tf, rlocus, bode, nyquist
```

```python
G = tf([0, 0, 1], [1, 1, 0])
```

```python
real, imag, freq = nyquist(G)
```

```python
mag, phase, omega = bode(G, logspace(-1, 2, 100), dB=True)
```

```python
r, k = rlocus(G.feedback(tf([1], [1])))
```
