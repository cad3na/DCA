```python
%matplotlib inline
from matplotlib.pyplot import plot, figure
```

```python
from control import tf, bode, nyquist, rlocus
```

```python
from numpy import log10, logspace, linspace
```

Trazar el diagrama de Bode de:

$$
G(s) =
    \frac{9(s^2 + 0.2 s + 1)}{s(s^2 + 1.2s + 9)}
$$

El termino constante, 9 se ve en:

$|K|_{dB} = 20 \log{9}$

```python
20*log10(9)
```

es decir $19 dB$.

```python
figure(figsize=(8, 6))
constante = tf([9],[1])
mag, pahse, omega = bode(constante, omega = logspace(-2, 3, 100), dB=True)
```

Para el termino integral $\frac{1}{s}$:

```python
figure(figsize=(8, 6))
integral = tf([0, 1],[1, 0])
mag, phase, omega = bode(integral, omega = logspace(-2, 3, 100), dB=True)
```

Para el denominador de segundo orden:

```python
figure(figsize=(8, 6))
denominador = tf([0,0,1],[1, 1.2, 9])
mag, phase, omega = bode(denominador, omega = logspace(-2, 3, 100), dB=True)
```

Para el numerador de segundo orden:

```python
figure(figsize=(8, 6))
numerador = tf([1, 0.2, 1], [0,0,1])
mag, phase, omega = bode(numerador, omega = logspace(-2, 3, 100), dB=True)
```

La respuesta de todo el sistema estaria dada por:

```python
figure(figsize=(8, 6))
mag, phase, omega = bode(constante*numerador*denominador*integral, omega = logspace(-2, 3, 100), dB=True)
```

Asi pues, un diagrama de Bode de un sistema, se puede ver como una superposición de todos sus elementos básicos.

```python
figure(figsize=(8, 6))
real, imag, freq = nyquist(constante*numerador*denominador*integral, omega=linspace(0, 100, 1000))
```
