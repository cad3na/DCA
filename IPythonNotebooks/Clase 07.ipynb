# Compensador de adelanto de fase

Implementamos un controlador con función de trasferencia:

$$
G_C = \alpha \frac{Ts + 1}{\alpha Ts + 1}
$$

con $0 < \alpha < 1$.

Para este ejemplo tomamos $\alpha = 0.1$ y $T = 1$.

```python
a = 0.1
T = 1.0
Gc = tf([T*a, 1*a], [a*T, 1])
```

```python
f = figure(figsize=(10, 10))
nyquist(Gc)
ax = f.gca()
ax.set_xlim(0, 1)
ax.set_ylim(-0.5, 0.5);
```

```python
figure(figsize=(10, 10))
mag, phase, omega = bode(Gc, logspace(-1, 2, 100), dB=True)
```

```python
phi = degrees(arcsin((1-a)/(1+a)))
phi
```

```python
omega = 1/sqrt(a)
omega
```

Este compensador se utiliza para proporcionar un angulo de adelanto de fase suficiente para compensar el excesivo retardo de fase asociado con las componentes del sistema fijo.

# Ejercicio 9-1

Sea el sistema fijo:

$$
G(s) = \frac{4}{s(s+2)}
$$

Diseñar un compensador, tal que el error estatico de velocidad  $k_v = \lim_{s \to 0} s G_C(s) G(s)$  sea igual a  $20 s^{-1} = 20 Hz$, que el margen de fase sea al menos de  $50^o$  y el margen de ganancia sea al menos de  $10dB$.

## Solución

Se utilizará un compensador de adelanto de fase:

$$
G_C(s) = k_C \alpha \frac{Ts + 1}{\alpha Ts + 1}
$$

donde $0 < \alpha < 1$.

### Paso 1 - Calcular $k_v$

$$
k_v
    = \lim_{s \to 0} s G_C G
    = \lim_{s \to 0} s \left( k_C \alpha \frac{Ts + 1}{\alpha Ts + 1} \right) \left( \frac{4}{s(s+2)} \right)
    = 2 k_C \alpha = 20
$$

$$
k_C \alpha = 10 = k
$$

### Paso 2 - Calcular el margen de fase $\gamma$ y el margen de ganancia $K_g$

```python
# Se define la funcion de transferencia del sistema 
G = tf([0, 4], [1, 0])*tf([0, 1], [1, 2])
```

```python
# Se le pone una ganancia positiva "k" al sistema
k = 10
G1 = tf([k], [1])*G
```

```python
# Se realiza la grafica de Bode
figure(figsize=(10, 10))
mag, phase, omega = bode(G1, logspace(-1, 2, 100), dB=True)
```

```python
# Se calculan el margen de ganancia Kg y la frecuencia a la
# que ocurre omega_g, asi como el margen de fase gamma y la
# frecuencia a la que ocurre omega_gamma
Kg, gamma, omega_g, omega_gamma = margin(mag, phase, omega)
print "El margen de ganancia es: " + str(Kg) + " y ocurre a la frecuencia: " + str(omega_g)
print "El margen de fase es: " + str(gamma) + " y ocurre a la frecuencia: " + str(omega_gamma)
```

### Paso 3 - Calcular el desfasamiento maximo $\phi_{max}$ y $\alpha$

Dentro de las especificaciones del controlador, teniamos que el margen de fase deberia ser mayor a $50^o$, por lo que tenemos que incrementarlo con un $\phi_{max}$ de:

$$
\phi_{max} \ge 50^o - \gamma = 50^o - 19.26^o = 30.73
$$

```python
50 - gamma
```

Podemos escoger uno mayor, por ejemplo $\phi_{max} = 32^o$.

De esta manera tenemos la siguiente relación con $\alpha$:

$$
\sin{\phi_{max}} = \frac{1-\alpha}{1+\alpha} \quad \implies \quad \alpha = \frac{1 - \sin{\phi_{max}}}{1 + \sin{\phi_{max}}}
$$

```python
a = (1 - sin(32))/(1 + sin(32))
a
```

Por lo que tenemos que $\alpha = 0.28$.

### Paso 4 - Ajustar la frecuencia critica $\omega_C$ para obtener $\phi_{max}$

Sabemos la relación entre la frecuencia de critica $\omega_C$ y $\phi_{max}$:

$$
20 \log{\frac{1}{\sqrt{\alpha}}} = -20 \log{\sqrt{\alpha}} = -10 log{\alpha} = 5.38 dB
$$

```python
-10*log10(a)
```

Por lo que ahora debemos buscar una frecuencia $\omega$ para que cuando sea evaluada en $\left| G_1(j \omega) \right|$ nos de una magnitud de $-5.38 dB$; la cual será $\omega_C$.

```python
omega_c = 8.5
20*log10(abs(G1.evalfr(omega_c)[0][0]))
```

Como podemos ver, la frecuencia de $8.5 \frac{rad}{s}$ se acerca lo suficiente a lo que queriamos obtener.

### Paso 5 - Calcular el compensador

Una vez que tenemos todos estos parametros, podemos calcular facilmente el compensador:

$$
T = \frac{1}{\sqrt{\alpha}{\omega_C}} = 0.2187
$$

```python
T = 1/(sqrt(a)*omega_c)
T
```

$$
\frac{1}{T} = 4.57
$$

```python
1/T
```

$$
\alpha T = 0.0632
$$

```python
a*T
```

$$
\frac{1}{\alpha T} = 15.807
$$

```python
1/(a*T)
```

$$
k_C = \frac{k}{\alpha} = 34.585
$$

```python
kc = k/a
kc
```

Y ahora podemos definir la función de transferencia del controlador como:

$$
G_C(s) = 10 \frac{0.218 s + 1}{0.063 s +1}
$$

```python
Gc = tf([T, 1], [a*T, 1])*tf([k],[1])
print Gc
```

Si trazamos el diagrama de Bode de todo el sistema, tendremos:

```python
figure(figsize=(10, 10))
mag, phase, omega = bode(Gc*G, logspace(-1, 2, 100), dB=True)
```

```python
# Se calculan el margen de ganancia Kg y la frecuencia a la
# que ocurre omega_g, asi como el margen de fase gamma y la
# frecuencia a la que ocurre omega_gamma
Kg, gamma, omega_g, omega_gamma = margin(mag, phase, omega)
print "El margen de ganancia es: " + str(Kg) + " y ocurre a la frecuencia: " + str(omega_g)
print "El margen de fase es: " + str(gamma) + " y ocurre a la frecuencia: " + str(omega_gamma)
```

Si queremos ver el comportamiento del sistema sin compensador, con una ganancia proporcional y con el compensador, al mismo tiempo, podemos hacer lo siguiente:

```python
figure(figsize=(10, 10))
mag, phase, omega = bode(G, logspace(-1, 2, 100), dB=True)
mag, phase, omega = bode(G1, logspace(-1, 2, 100), dB=True)
mag, phase, omega = bode(Gc*G, logspace(-1, 2, 100), dB=True)
```

Y para la respuesta en tiempo del sistema:

```python
Gf = G.feedback(tf([1],[1]))
G1f = G1.feedback(tf([1],[1]))
Gtf = (G*Gc).feedback(tf([1],[1]))
```

```python
tf, yf = step_response(Gf, T=linspace(0, 10, 1000))
t1, y1 = step_response(G1f, T=linspace(0, 10, 1000))
tt, yt = step_response(Gtf, T=linspace(0, 10, 1000))
```

```python
figure(figsize=(10, 10))
p1, = plot(tf, yf)
p2, = plot(t1, y1)
p3, = plot(tt, yt)
legend([p1, p2, p3], [r"$G_F$", r"$k G_F$", "$G_c G_F$"]);
```

# Ejemplo 9-2

```python
K = 5.0
```

```python
G = tf([0, 1], [1, 0]) * tf([0, 1], [1, 1]) * tf([0, 1], [0.5, 1])
```

```python
G1 = tf([K], [1])*G
```

```python
figure(figsize=(10, 10))
mag, phase, omega = bode(G1)
```

```python
margin(mag, phase, omega)
```

```python
b = 10
T = 10
```

```python
Gc = tf([T, 1], [b*T, 1])
```

```python
figure(figsize=(10, 10))
mag, phase, omega = bode(G, logspace(-1, 2, 100), dB=True)
mag, phase, omega = bode(G1, logspace(-1, 2, 100), dB=True)
mag, phase, omega = bode(Gc*G1, logspace(-1, 2, 100), dB=True)
```

```python
Gf = G.feedback(tf([1],[1]))
G1f = G1.feedback(tf([1],[1]))
Gtf = (G1*Gc).feedback(tf([1],[1]))
```

```python
tf, yf = step_response(Gf, T=linspace(0, 30, 1000))
t1, y1 = step_response(G1f, T=linspace(0, 30, 1000))
tt, yt = step_response(Gtf, T=linspace(0, 30, 1000))
```

```python
f = figure(figsize=(10, 10))
plot(tf, yf, label="Gf")
plot(t1, y1, label="G1f")
plot(tt, yt, label="Gtf")
ax = f.gca()
ax.set_ylim(-0.2, 1.5);
```
