```python
from numpy import mgrid, sqrt
```

```python
%matplotlib inline
from matplotlib.pyplot import streamplot, cm, figure, style
```

```python
Y, X = mgrid[-10:10:100j, -10:10:100j]
U = X + 2*Y
V = 3*X + 2*Y
speed = sqrt(U*U + V*V)
```

```python
f = figure(figsize=(7, 7))
style.use("ggplot")
streamplot(X, Y, U, V, color=U, linewidth=2, cmap=cm.cool)
ax = f.gca()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10);
```

```python
f = figure(figsize=(7, 7))
lw = 5*speed/speed.max()
style.use("ggplot")
streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)
ax = f.gca()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10);
```
