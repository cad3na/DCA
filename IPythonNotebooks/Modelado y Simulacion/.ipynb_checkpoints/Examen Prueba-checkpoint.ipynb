# Ejercicio 1

Conseidere la dinámica de una población de peces descrita por medio del conocido método de Verhulst:

$$
\dot{x}(t) = r x(t) \left( 1 - \frac{x(t)}{B} \right)
$$

donde $x$ denota la biomasa, y $r$ y $B$ con parámetros constantes y positivos. Suponga que se desea capturar peces de acuerdo con la ley de captura dada por:

$$
u(t) = \alpha x^2(t)
$$

con $\alpha$ positiva y constante. Analice las circunstancias de equilibrio del sistema sometido a la pesca y determine las condiciones sobre los parametros que garantizan la fisicalidad del sistema. Tambien obtenga la ecuación gobernante del sistema sometido a la pesca.

## Solución

```python
%matplotlib inline
from matplotlib.pyplot import plot, figure, style
from numpy import linspace, array
from scipy.integrate import odeint
from IPython.html.widgets import interact, fixed
```

Dado el sistema descrito en el problema, cuando se agrega la condición de pesca a un modelo de crecimiento logistico tendremos:

$$
\dot{x}(t) = r x(t) \left( 1 - \frac{x(t)}{B} \right) - u(t)
$$

es decir:

$$
\dot{x}(t) = r x(t) \left( 1 - \frac{x(t)}{B} \right) - \alpha x^2(t)
$$

por lo que el modelo queda como sigue:

$$
\dot{x} = r x(t) \left( 1 - x(t) \left( \frac{1}{B} + \alpha \right) \right)
$$

de donde podemos identificar que $r$ y $\frac{1}{B} + \alpha$ son parametros constantes y positivos.

Por otro lado, sabemos que los puntos de equilibrio del sistema, son aquellos que hacen a $\dot{x} = 0$, por lo que si graficamos la relación entre $\dot{x}$ y $x$, podremos obtener estos puntos por una inspección rapida:

```python
def xp(x, r, B, a):
    """Regresa el comportamiento del sistema dados los parametros."""
    return r*x*(1 - x*(1/B + a))
```

```python
def compsistema(sistema, X, r=0.1, B=2.0, a=0.2):
    """Imprime el comportamiento del sistema, dado el sistema, un vector de estados de entrada y los parametros del sistema"""
    XP = sistema(X, r, B, a)
    
    f = figure(figsize=(14, 7))
    style.use("ggplot")
    plot(X, XP)
    
    ax = f.gca()
    ax.set_xlim(-0.1*X.max(), X.max()*1.2)
    ax.set_ylim(-0.1*XP.max(), XP.max()*1.2)
    ax.set_xlabel(r"$x(t)$", fontsize=18)
    ax.set_ylabel(r"$\dot{x}(t)$", fontsize=18);
```

```python
interact(compsistema, sistema=fixed(xp), X=fixed(linspace(-1, 2, 1000)), r=(0.0, 10.0), B=(0.0, 10.0), a=(0.0, 10.0))
```

Por la manera en que esta construido el modelo, este sistema siempre tendrá un punto de equilibrio en el origen, por lo que podemos notar que la solución trivial siempre se encontrará. Por otro lado, al inspeccionar las condiciones de equilibrio se obtiene que el otro punto de equilibrio se obtiene cuando:

$$
x(t) = \frac{r B}{r + B \alpha}
$$

Si sondeamos la solución a la ecuación diferencial de un conjunto fijo de parametros, obtendremos:

```python
xp1 = lambda x, t: xp(x, r=0.1, B=2.0, a=0.2)
```

```python
def imprimesalida(sistema, x0=2.0, tmax=10):
    
    t = linspace(0, tmax, 1000)
    salida = odeint(sistema, x0, t)
    
    f = figure(figsize=(14, 7))
    style.use("ggplot")
    plot(t, salida)
    ax = f.gca()
    ax.set_xlabel(r"$t$", fontsize=18)
    ax.set_ylabel(r"$x(t)$", fontsize=18);
```

```python
interact(imprimesalida, sistema=fixed(xp1), x0=(0.0, 10.0), tmax=(0.0, 20.0));
```

Modifique el modelo Lotka-Volterra para introducir la muerte natural de las presas y de los depredadores por causas del envejecimiento. Justifique el modelo propuesto en términos analíticos (circunstancias de equilibrio, fisicalidad, etc.).

## Solución

```python
from numpy import mgrid, sqrt
from matplotlib.pyplot import streamplot, cm
```

```python
f = figure(figsize=(7, 7))
style.use("ggplot")
streamplot(x1, x2, dx1, dx2, color=dx1, linewidth=2, cmap=cm.cool)
ax = f.gca()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5);
```

```python
a, b, c, d, e = 1.0, 2.0, 2.5, 1.5, 0.9

x2, x1 = mgrid[-5:5:1000j, -5:5:1000j]
dx1 = (a - e)*x1  - b* x1*x2
dx2 = (-c - e)*x2 + d*x1*x2
```

```python
def predep(X, t=0):
    x1, x2 = X
    a, b, c, d, e = 1.0, 2.0, 2.5, 1.5, 0.9
    dx1 = (a - e)*x1  - b* x1*x2
    dx2 = (-c - e)*x2 + d*x1*x2
    return dx1, dx2
```

```python
def planofase(sistema, x1_0, x2_0, x1_max=7.0, x2_max=7.0, t_max=10):
    
    x2, x1 = mgrid[-0.1*x2_max:x2_max*1.2:100j, -0.1*x1_max:x1_max*1.2:100j]
    dx1, dx2 = sistema([x1, x2])
    speed = sqrt(dx1*dx1 + dx2*dx2)
    
    x0 = array([x1_0, x2_0])
    t = linspace(0, t_max, 1000)
    
    respuesta = odeint(sistema, x0, t)
    
    f = figure(figsize=(x1_max, x2_max))
    lw = 10*speed/speed.max()
    style.use("ggplot")
    streamplot(x1, x2, dx1, dx2, density=0.6, color='k', linewidth=lw)
    plot(respuesta[:, 0], respuesta[:,1])
    ax = f.gca()
    ax.set_xlim(-0.1*x1_max, 1.2*x1_max)
    ax.set_ylim(-0.1*x2_max, 1.2*x2_max);
```

```python
planofase(predep, 3, 0.1)
```
