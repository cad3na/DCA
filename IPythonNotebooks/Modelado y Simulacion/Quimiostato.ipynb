Dado el sistema no lineal:

$$
\frac{dN}{dt} = \alpha_1 \frac{C}{1 + C} N -  N
$$

$$
\frac{dC}{dt} = - \frac{C}{1 + C} N - C + \alpha_2
$$

donde la variable de estado $x$ esta definida por:

$$
x =
\begin{pmatrix}
N \\
C
\end{pmatrix}
$$

entonces el sistema descrito, se puede simplificar por lo siguiente:

$$
\dot{x} = f(x)
$$

sabemos, por un análisis anterior, que los puntos de equilibrio son:

$$
x_1 =
\begin{pmatrix}
0 \\
\alpha_2
\end{pmatrix}
$$

$$
x_2 =
\begin{pmatrix}
\frac{1}{\alpha_1 - 1} \\
\alpha_1 \alpha_2 - \frac{\alpha_1}{\alpha_1 - 1}
\end{pmatrix}
$$

```python
%matplotlib inline
from matplotlib.pyplot import streamplot, cm, figure, style, plot
from numpy import mgrid, sqrt, array, linspace
from scipy.integrate import odeint
from IPython.html.widgets import interact, fixed
```

```python
a1, a2 = 2.0, 2.0

C, N = mgrid[-5:5:100j, -5:5:100j]
dN = a1*(C/(1 + C))*N  - N
dC = -(C/(1 + C))*N - C + a2
speed = sqrt(dN*dN + dC*dC)
```

```python
f = figure(figsize=(7, 7))
style.use("ggplot")
streamplot(N, C, dN, dC, color=dC, linewidth=2, cmap=cm.cool)
ax = f.gca()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5);
```

```python
f = figure(figsize=(7, 7))
lw = 15*speed/speed.max()
style.use("ggplot")
streamplot(N, C, dN, dC, density=0.6, color='k', linewidth=lw)
ax = f.gca()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5);
```

```python
def quimiostato(X, t=0):
    """Esta función regresa el comportamiento de un quimiostato dad"""
    N = X[0]
    C = X[1]
    a1 = 2.0
    a2 = 2.0
    
    dN = a1*(C/(1 + C))*N  - N
    dC = -(C/(1 + C))*N - C + a2
    
    return dN, dC
```

```python
def compsistema(sistema, C0, N0, Cmax=8, Nmax=8, tmax=10):

    C, N = mgrid[-1*Cmax:Cmax:100j, -1*Nmax:Nmax:100j]
    dN, dC = sistema(array([N, C]))
    speed = sqrt(dN*dN + dC*dC)

    x0 = array([C0, N0])
    t = linspace(0, tmax, 1000)

    respuesta = odeint(sistema, x0, t)

    f = figure(figsize=(Cmax, Nmax))
    lw = 30*speed/speed.max()
    style.use("ggplot")
    streamplot(N, C, dN, dC, density=0.6, color='k', linewidth=lw)
    plot(respuesta[:, 0], respuesta[:,1])
    ax = f.gca()
    ax.set_xlim(-Cmax, Cmax)
    ax.set_ylim(-Nmax, Nmax);
```

```python
compsistema(quimiostato, 2, 3)
```

```python
interact(compsistema, sistema=fixed(quimiostato), C0=(0.0, 5.0), N0=(0.0, 5.0))
```
