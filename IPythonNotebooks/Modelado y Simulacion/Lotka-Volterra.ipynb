```python
from numpy import mgrid, sqrt, e
```

```python
%matplotlib inline
from matplotlib.pyplot import streamplot, cm, figure, style
```

```python
a, b, c, d = 1, 2, 3, 4

x2, x1 = mgrid[-10:10:1000j, -10:10:1000j]
dx1 = a*x1  - b* x1*x2
dx2 = -c*x2 + d*x1*x2
speed = sqrt(dx1*dx1 + dx2*dx2)
```

```python
f = figure(figsize=(7, 7))
style.use("ggplot")
streamplot(x1, x2, dx1, dx2, color=dx1, linewidth=2, cmap=cm.cool)
ax = f.gca()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10);
```

```python
f = figure(figsize=(7, 7))
lw = 5*speed/speed.max()
style.use("ggplot")
streamplot(x1, x2, dx1, dx2, density=0.6, color='k', linewidth=lw)
ax = f.gca()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10);
```
