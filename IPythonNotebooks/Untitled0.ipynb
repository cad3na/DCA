```python
%matplotlib inline
from matplotlib.pyplot import plot, figure, gca, legend
```

```python
from control import ss, lsim
from numpy import matrix, sin, linspace
```

```python
# Volvemos a definir las matrices, pero como arreglos de numpy
# debido a que el modulo de control no sabe que hacer con las
# variables simbolicas
A = matrix([[0, 1], [0, -2]])
b = matrix([[0], [1]])
c = matrix([[4], [0]])
d = 0
f = matrix([[-16], [-2]])
Af = A + b*f.T
cf = c + d*f
# Se agrega una ganancia de 4 a la entrada del sistema, debido
# a que crece la norma del denominador, y por la tanto se atenua
# la salida del sistema.
SR = ss(Af, 4*b, cf.T, d)
```

```python
t = linspace(0, 10, 1000)
u = 1.0
y, t, x = lsim(SR, U=u, T=t)
```

```python
figure(figsize=(18, 10))
p1, p2, = plot(t, x)
p3, = plot(t, y)
legend([p1, p2, p3], [r"$x(t)$", r"$x'(t)$", r"$y(t)$"]);
```
