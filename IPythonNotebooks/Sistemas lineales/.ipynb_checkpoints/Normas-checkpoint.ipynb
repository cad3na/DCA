```python
from numpy import linspace
```

```python
norma1 = lambda x, y: x + y
norma2 = lambda x, y: (x**2 + y**2)**(1.0/2)
norma3 = lambda x, y: (x**3 + y**3)**(1.0/3)
```

```python
norma1(1, 2)
```

```python
norma2(1.0, 2.0)
```

```python
f1 = lambda x: 1 - x
f1_25 = lambda x: (1 - x**(1.25))**(1.0/1.25)
f1_5 = lambda x: (1 - x**(1.5))**(1.0/1.5)
f2 = lambda x: (1 - x**2)**(1.0/2)
f3 = lambda x: (1 - x**3)**(1.0/3)
f5 = lambda x: (1 - x**5)**(1.0/5)
f10 = lambda x: (1 - x**10)**(1.0/10)
f100 = lambda x: (1 - x**100)**(1.0/100)
```

```python
x = linspace(0, 1, 1000)
```

```python
%matplotlib inline
```

```python
from matplotlib.pyplot import style, plot, figure, legend
```

```python
style.use("ggplot")
```

```python
f = figure(figsize=(7, 7))

p1, = plot(x, f1(x))
p1_25, = plot(x, f1_25(x))
p1_5, = plot(x, f1_5(x))
p2, = plot(x, f2(x))
p3, = plot(x, f3(x))
p5, = plot(x, f5(x))
p10, = plot(x, f10(x))
p100, = plot(x, f100(x))

ax = f.gca()
ax.set_xlim(0, 1.1)
ax.set_ylim(0, 1.1);

legend([p1, p1_25, p1_5, p2, p3, p5, p10, p100], [r"$||x||_1$", r"$||x||_{1.25}$",
                                                 r"$||x||_{1.5}$", r"$||x||_2$", r"$||x||_3$",
                                                 r"$||x||_5$", r"$||x||_{10}$", r"$||x||_{100}$"], loc=3);
```
