```python
from control import step_response, ss
```

```python
from numpy import matrix, linspace
```

```python
%matplotlib inline
from matplotlib.pyplot import plot, figure, gca, legend, style
```

```python
A = matrix([[0, 1, 0], [0, 0, 1], [0, 0, 0]])
b = matrix([[0], [0], [1]])
c = matrix([[-1], [0], [1]])
x0 = matrix([[1], [0], [0]])
```

```python
SR = ss(A, b, c.T, 0)
```

```python
t, y = step_response(SR, T=linspace(0, 10, 1000))
```

```python
figure(figsize=(15, 6))
style.use("ggplot")
plot(t, y)
```

Si usamos retroalimentación de estado:

```python
f1 = matrix([[1], [1], [-1]])
```

```python
Af1 = A + b*f1.T
```

```python
SR1 = ss(Af1, b, c.T, 0)
t, y = step_response(SR1, T=linspace(0, 10, 1000))
```

```python
figure(figsize=(15, 6))
style.use("ggplot")
plot(t, y)
```
