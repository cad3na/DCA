```python
%matplotlib inline
```

```python
from matplotlib.pyplot import plot, figure, gca, legend, style, subplot, subplots
```

```python
import scipy.io
adaptableGradiente = scipy.io.loadmat('./MATLAB/adaptableGradiente.mat')
```

```python
salidas = adaptableGradiente.get("salidas")
```

```python
control = adaptableGradiente.get("control")
```

```python
t = adaptableGradiente.get("tout")
```

```python
a1 = adaptableGradiente.get("datosa1")
b1 = adaptableGradiente.get("datosb1")
b0 = adaptableGradiente.get("datosb0")
```

```python
referencia = salidas[:, 0]
salida = salidas[:, 1]
```

```python
f = figure(figsize=(18, 7))
style.use("ggplot")
p1, = plot(t, salida)
p2, = plot(t, referencia)

ax = f.gca()
ax.set_ylim(-1.5, 1.5);

legend([p1, p2], [r"$y(t)$", r"$r(t)$"]);
```

```python
f = figure(figsize=(18, 7))
style.use("ggplot")
p1, = plot(t, control)

ax = f.gca()
ax.set_ylim(-1.5, 1.5);

legend([p1], [r"$u(t)$"]);
```

```python
f = figure(figsize=(18, 7))
style.use("ggplot")
subplot(1, 3, 1)
p1, = plot(t, a1)
ax = f.gca()
ax.set_ylim(-0.1, 0.6);
legend([p1], [r"$\hat{a}_1$"]);

subplot(1, 3, 2)
p1, = plot(t, b1)
ax = f.gca()
ax.set_ylim(-8.0, -5);
legend([p1], [r"$\hat{b}_1$"]);

subplot(1, 3, 3)
p1, = plot(t, b0)
ax = f.gca()
ax.set_ylim(0.4, 1.2);
legend([p1], [r"$\hat{b}_0$"]);
```

```python
import scipy.io
adaptableGradiente = scipy.io.loadmat('./MATLAB/adaptableGradiente.mat')
```

```python
salidas = adaptableGradiente.get("salidas")
```

```python
referencia = salidas[:, 0]
salida = salidas[:, 1]
```

```python
control = adaptableGradiente.get("control")
```

```python
t = adaptableGradiente.get("tout")
```

```python
parametros = adaptableGradiente.get("parametros")
p1 = parametros[:, 1]
p2 = parametros[:, 2]
p3 = parametros[:, 3]
p4 = parametros[:, 4]
p5 = parametros[:, 5]
p6 = parametros[:, 6]
```

```python
f = figure(figsize=(18, 7))
style.use("ggplot")
pl1, = plot(t, salida)
pl2, = plot(t, referencia)

ax = f.gca()
ax.set_ylim(-1.5, 1.5);

legend([pl1, pl2], [r"$y(t)$", r"$r(t)$"]);
```

```python
f = figure(figsize=(18, 7))
style.use("ggplot")
p1, = plot(t, control)

ax = f.gca()
ax.set_ylim(-1.5, 1.5);

legend([p1], [r"$u(t)$"]);
```

```python
t = parametros[:, 0]
fig, axes = subplots(ncols=3, nrows=2)
ax1, ax2, ax3, ax4, ax5, ax6 = axes.ravel()
fig.set_size_inches(18,14)
style.use("ggplot")

pl1, = ax1.plot(t, p1, label=r"$p_1$")
ax1.legend()

pl2, = ax2.plot(t, p2, label=r"$p_2$")
ax2.legend()

pl3, = ax3.plot(t, p3, label=r"$p_3$")
ax3.legend()

pl4, = ax4.plot(t, p4, label=r"$p_4$")
ax4.legend()

pl5, = ax5.plot(t, p5, label=r"$p_5$")
ax5.legend()

pl6, = ax6.plot(t, p6, label=r"$p_6$")
ax6.legend();
```
